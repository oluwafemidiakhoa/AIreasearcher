````python
import asyncio
import logging
import nest_asyncio
from concurrent.futures import ThreadPoolExecutor
import spacy
from Bio import Entrez
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
from retry import retry
from smolagents import CodeAgent, LiteLLMModel
from google.colab import userdata

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Load SpaCy model and configure Entrez
nlp = spacy.load("en_core_web_sm")
Entrez.email = "oluwafemidiakhoa@gmail.com"  # Replace with your email

# Initialize SmolAgents LLM model
model = LiteLLMModel(model_id="gpt-4o", api_key=userdata.get("OPENAI_API_KEY"))  # Replace with your API key

# Set up a thread pool executor for blocking operations
executor = ThreadPoolExecutor(max_workers=5)

def enforce_code_block_format(text):
    """Ensure text is wrapped in a Python code block."""
    if isinstance(text, dict):
        return text  # Skip processing if already a dictionary
    if not isinstance(text, str):
        logger.error("Expected text to be a string or dictionary.")
        return "Error: Invalid text format."
    if not text.startswith("```"):
        return f"```python\n{text}\n```"
    return text

@retry(tries=3, delay=2, backoff=2, logger=logger)
def fetch_pubmed_articles_sync(query, max_results=5):
    """Fetch articles from PubMed with retry logic."""
    try:
        handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results)
        record = Entrez.read(handle)
        handle.close()
        article_ids = record['IdList']
        articles = []
        for article_id in article_ids:
            fetch_handle = Entrez.efetch(db="pubmed", id=article_id, rettype="abstract", retmode="text")
            articles.append(fetch_handle.read())
            fetch_handle.close()
        return articles
    except Exception as e:
        logger.error(f"Error fetching articles: {e}")
        return []

async def fetch_pubmed_articles(query, max_results=5):
    """Fetch articles asynchronously using ThreadPoolExecutor."""
    loop = asyncio.get_event_loop()
    articles = await loop.run_in_executor(executor, fetch_pubmed_articles_sync, query, max_results)
    return articles

def extract_keywords(text, max_keywords=10):
    """Extract keywords from text."""
    if not isinstance(text, str):
        text = str(text)
    doc = nlp(text)
    keywords = [token.text for token in doc if token.is_alpha and not token.is_stop and token.pos_ in {"NOUN", "VERB"}]
    return keywords[:max_keywords]

def summarize_article(article):
    """Summarize the article using CodeAgent."""
    try:
        if not isinstance(article, str):
            article = str(article)
        agent = CodeAgent(model=model, tools=[])
        prompt = f"Summarize the following article in 100 words or less:\n\n{article}"
        summary = agent.run(prompt)
        if isinstance(summary, dict):
            summary = summary.get("Summary", "No summary available.")
        return summary.strip()
    except Exception as e:
        logger.error(f"Error summarizing article: {e}")
        return "Error generating summary."

def compute_similarity(summaries):
    """Compute semantic similarity between summaries."""
    if not summaries:
        logger.warning("No summaries available for similarity computation.")
        return np.array([])
    embeddings = [nlp(str(summary)).vector for summary in summaries]
    similarity_matrix = cosine_similarity(embeddings, embeddings)
    return similarity_matrix

def visualize_similarity_matrix(matrix, summaries):
    """Visualize the similarity matrix as a heatmap."""
    if matrix.size == 0:
        logger.warning("Empty similarity matrix, skipping visualization.")
        return
    plt.figure(figsize=(10, 8))
    plt.imshow(matrix, cmap='viridis', interpolation='nearest')
    plt.colorbar(label='Cosine Similarity')
    plt.xticks(ticks=np.arange(len(summaries)), labels=[f"Summary {i+1}" for i in range(len(summaries))], rotation=45)
    plt.yticks(ticks=np.arange(len(summaries)), labels=[f"Summary {i+1}" for i in range(len(summaries))])
    plt.title("Semantic Similarity Matrix")
    plt.tight_layout()
    plt.show()

def generate_insights(summaries, similarity_matrix):
    """Generate insights from summaries and similarity matrix."""
    if not summaries or similarity_matrix.size == 0:
        logger.warning("Insufficient data for insights generation.")
        return
    try:
        agent = CodeAgent(model=model, tools=[])
        insights_prompt = f"""
        Analyze the following summaries and their semantic similarities to identify:
        1. Overlapping themes.
        2. Unique findings in each article.
        3. Key challenges and future directions.

        Summaries:
        {summaries}

        Similarity Matrix:
        {similarity_matrix}

        Provide your insights using the following format:

        ```python
        # Insights in Python code
        def insights():
            return {{
                "Overlapping Themes": "...",
                "Unique Findings": "...",
                "Key Challenges and Future Directions": "..."
            }}
        ```
        """
        insights = agent.run(insights_prompt)
        insights = enforce_code_block_format(insights)
        logger.info("\n--- Automated Insights ---")
        logger.info(insights)
    except Exception as e:
        logger.error(f"Error generating insights: {e}")

def log_processing_step(step_description):
    """Log each processing step in detail."""
    logger.info(f"Step: {step_description}")

async def async_multi_agent_workflow(query, max_results=5):
    """Asynchronous multi-agent workflow."""
    log_processing_step("Fetching articles asynchronously...")
    articles = await fetch_pubmed_articles(query, max_results)
    if not articles:
        logger.warning("No articles fetched. Aborting workflow.")
        return []
    results = []
    summaries = []
    for i, article in enumerate(articles):
        log_processing_step(f"Processing Article {i+1}")
        summary = summarize_article(article)
        summaries.append(summary)
        keywords = extract_keywords(article)
        results.append({"summary": summary, "keywords": keywords, "article": article})
    similarity_matrix = compute_similarity(summaries)
    log_processing_step("Computing semantic similarity matrix.")
    logger.info(f"Semantic Similarity Matrix:\n{similarity_matrix}")
    visualize_similarity_matrix(similarity_matrix, summaries)
    log_processing_step("Generating insights.")
    generate_insights(summaries, similarity_matrix)
    return results

async def main():
    """Main function to execute the workflow."""
    research_query = "AI applications in cancer diagnosis"
    log_processing_step("Starting asynchronous multi-agent workflow...")
    results = await async_multi_agent_workflow(research_query, max_results=5)
    log_processing_step("Workflow completed.")
    return results

if __name__ == "__main__":
    asyncio.run(main())
````
